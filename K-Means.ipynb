```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from mpl_toolkits.mplot3d import Axes3D
```

```python
img = cv.imread('indianrailways.jpeg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
fig, ax = plt.subplots(1, figsize=(12, 8))
ax.imshow(img)
plt.show()
```

```python
img_k = img.reshape((-1, 3))
img_k = np.float32(img_k)
criteria = (cv.TERM_CRITERIA_MAX_ITER,50, 1.0)
flags = cv.KMEANS_PP_CENTERS
compactness,labels,centers = cv.kmeans(img_k, 5, None, criteria, 10, flags)
```

```python
centers = np.uint8(centers)
centers.shape
```

```python
res = centers[labels.flatten()]
result_image = res.reshape((img.shape))
```

```python
fig, ax = plt.subplots(1, figsize=(12, 8))
# ax.imshow(img)
ax.imshow(result_image, alpha=0.6)
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
ax = Axes3D(fig)
ax.scatter(centers[:, 0], centers[:, 1], centers[:, 2], c='k', s=100)
ax.scatter(img_k[:,0], img_k[:,1], img_k[:, 2], alpha=0.1)
# ax.view_init(180)
plt.show()
```
